```python
import json
import pprint as pp
import pandas as pd
import tqdm
```

```python
def load_declarations():
    with open('declarations.json', 'r') as f:
        distros_dict = json.load(f)
    return distros_dict
```

```python
declarations = load_declarations()
```

```python
pp.pprint(declarations[6])
```

```python
persons = {}
```

```python
for declaration in tqdm.tqdm(declarations):
    pid = declaration["main"]["person"]["id"]
    if pid in persons:
        persons[pid]["sections"].append(declaration)
    else:
        new_person = {}
        new_person['family_name'] = declaration["main"]["person"]["family_name"]
        new_person['given_name'] = declaration["main"]["person"]["given_name"]
        new_person['name'] = declaration["main"]["person"]["name"]
        new_person['patronymic_name'] = declaration["main"]["person"]["patronymic_name"]
        new_person['id'] = pid
        new_person['sections'] = [declaration]
        
        persons[pid] = new_person
```
